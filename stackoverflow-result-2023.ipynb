```python
# Calculate statistics for the 'ConvertedCompYearly' column
mean_compensation = df['ConvertedCompYearly'].mean()
std_dev_compensation = df['ConvertedCompYearly'].std()
median_compensation = df['ConvertedCompYearly'].median()
min_compensation = df['ConvertedCompYearly'].min()
max_compensation = df['ConvertedCompYearly'].max()

# Calculate statistics for the 'WorkExp' column
mean_work_experience = df['WorkExp'].mean()
std_dev_work_experience = df['WorkExp'].std()
median_work_experience = df['WorkExp'].median()
min_work_experience = df['WorkExp'].min()
max_work_experience = df['WorkExp'].max()

# Print the computed statistics for 'ConvertedCompYearly'
print("Statistics for 'ConvertedCompYearly':")
print(f"Mean Compensation: {mean_compensation}")
print(f"Standard Deviation Compensation: {std_dev_compensation}")
print(f"Median Compensation: {median_compensation}")
print(f"Minimum Compensation: {min_compensation}")
print(f"Maximum Compensation: {max_compensation}")

# Print the computed statistics for 'WorkExp'
print("Statistics for 'WorkExp':")
print(f"Mean Work Experience: {mean_work_experience}")
print(f"Standard Deviation Work Experience: {std_dev_work_experience}")
print(f"Median Work Experience: {median_work_experience}")
print(f"Minimum Work Experience: {min_work_experience}")
print(f"Maximum Work Experience: {max_work_experience}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract and count the values from the 'LanguageHaveWorkedWith' and 'LanguageWantToWorkWith' columns
have_worked_with = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()
want_to_work_with = df['LanguageWantToWorkWith'].str.split(';', expand=True).stack().value_counts()

# Merge the counts into a single DataFrame
languages_df = pd.DataFrame({'HaveWorkedWith': have_worked_with, 'WantToWorkWith': want_to_work_with})
languages_df.fillna(0, inplace=True)  # Replace NaN values with 0

# Calculate the love percentages
languages_df['LovePercentage'] = (languages_df['WantToWorkWith'] / languages_df['HaveWorkedWith']) * 100

# Sort the data by 'LovePercentage' in descending order to show the most loved languages at the top
languages_df = languages_df.sort_values(by='LovePercentage', ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(languages_df.index, languages_df['LovePercentage'], color='skyblue')
plt.xlabel('Love Percentage')
plt.ylabel('Programming Language')
plt.title('Most Loved Programming Languages in the Stack Overflow 2023 Survey')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Count the number of respondents in each age group
age_counts = df['Age'].value_counts()

# Calculate the percentage for each age group
total_responses = len(df)
age_percentages = (age_counts / total_responses) * 100

# Create a bar chart for age group percentages
plt.figure(figsize=(10, 6))
age_percentages.plot(kind='bar', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Respondents')
plt.title('Percentage of Responses from Various Age Groups in the Stack Overflow 2023 Survey')
```
